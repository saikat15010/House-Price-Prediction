```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('train.csv')
dataset.shape
```

```python
dataset.head()
```

## Null values handling

```python
feature_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
len(feature_na)
```

```python
for feature in feature_na:
    print('{} have {}% null values'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

## Handling Null values in categorical variable

```python
feature_na1=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']
len(feature_na1)
```

```python
def replace_cat(dataset,feature_na1):
    data=dataset.copy()
    data[feature_na1]=data[feature_na1].fillna('null')
    return data
dataset=replace_cat(dataset,feature_na1)
dataset[feature_na1].isnull().sum()
```

```python
dataset.head()
```

## Handling Null values in numerical variable

```python
feature_na2=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']
len(feature_na2)
```

```python
for feature in feature_na2:
    print('{} have {}% null values'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
dataset[feature_na2].isnull().sum()
```

```python
for feature in feature_na2:
    
    median_value=dataset[feature].median()
    dataset[feature+'na']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)

dataset[feature_na2].isnull().sum()
```

```python
dataset.head()
```

## Temporal variable handling 

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()   ## Dataframe thats why [[]] for list it will be [].
```

```python
dataset.head()
```

## log normal distribution

```python
feature_log=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in feature_log:
    dataset[feature]=np.log(dataset[feature])
dataset.head()
```

## Handling rare category feature

```python
category_feature=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
```

```python
pd.pandas.set_option('display.max_rows',None)
```

```python
print(dataset.groupby('Neighborhood')['SalePrice'].count())
```

```python
len(dataset.Neighborhood.unique())
```

```python
for feature in category_feature:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
dataset.head(30)
```

```python
print(dataset.groupby('Neighborhood')['SalePrice'].count())
```

```python
len(dataset.Neighborhood.unique())
```

## Mean Encoding

```python
for feature in category_feature:
    label=dataset.groupby(feature)['SalePrice'].mean().sort_values().index
    dictionary={k:i for i,k in enumerate(label,0)}
    dataset[feature]=dataset[feature].map(dictionary)
```

```python
dataset.head(20)
```

## Feature Scalling

```python
feature_scale=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
feature_scale
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
sc.fit(dataset[feature_scale])
```

```python
x=sc.transform(dataset[feature_scale])
x
```

```python
dataset=pd.concat([dataset[['Id','SalePrice']],pd.DataFrame(x,columns=feature_scale)],axis=1)
dataset.head()
```

```python
dataset.to_csv('house.csv',index=False)
```
