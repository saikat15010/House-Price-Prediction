  ## In this section we will handle
  1. Null values
  2. Numerical values
  3. Discrete values
  4. Continous values
  5. Outlier

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
dataset.shape
```

```python
dataset.head()
```

## Section 1
## Check how many missing value present in the dataset

```python
feature_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
feature_na
```

```python
for feature in feature_na:
    print('{} have {}% missing values'.format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

## Now lets see if there any relation between the missing values and dependent variable

```python
for feature in feature_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature +' vs SalePrice')
    plt.show()
    
```

```python
## There is defintely relation between the nul values with dependent variable.
```

## Section 2
## Numerical Value

```python
feature_num=[feature for feature in dataset.columns if dataset[feature].dtypes!='O']
feature_num
```

```python
dataset[feature_num].head()
```

## 2.1- Temporal variable

```python
feature_temp=[feature for feature in feature_num if 'Yr' in feature or 'Year' in feature]
feature_temp
```

```python
for feature in feature_temp:
    print(feature,dataset[feature].unique())
```

```python
## Lets see the relation between the Yrsold with the Sale Price
```

```python
for feature in feature_temp:
    data=dataset.copy()
    if feature=='YrSold':
        data.groupby(feature)['SalePrice'].median().plot()
```

```python
## We can differentiate between YrSold with other temporal variable
```

```python
for feature in feature_temp:
    data=dataset.copy()
    if feature!='YrSold':
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

## 2.2- Discrete Value

```python
feature_dis=[feature for feature in feature_num if len(dataset[feature].unique())<25 and feature not in feature_temp+['Id']]
len(feature_dis)
```

```python
dataset[feature_dis].head()
```

```python
for feature in feature_dis:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature +' vs SalePrice')
    plt.show()
    
```

```python
## There must be relation between the discrete and dependent variable
```

## 2.3- Continous variable

```python
feature_con=[feature for feature in feature_num if feature not in feature_dis and feature not in feature_temp+['Id']]
len(feature_con)
```

```python
pd.pandas.set_option('display.max_rows',None)
```

```python
dataset[feature_con].head()
```

```python
for feature in feature_con:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
for feature in feature_con:
    data=dataset.copy()
    data[feature]=np.log(data[feature]+1)
    data['SalePrice']=np.log(data['SalePrice']+1)
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
for feature in feature_con:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

## Section 3
## Categorical Variable

```python
feature_cat=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
feature_cat
```

```python
dataset[feature_cat].head()
```

```python
len(feature_cat)
```

```python
for feature in feature_cat:
    print('{} have {} category'.format(feature,len(dataset[feature].unique())))
```

```python
for feature in feature_cat:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

## Outlier

```python
for feature in feature_con:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
